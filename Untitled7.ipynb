# End-to-End Climate Data Analysis

## 1. Data Loading and Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/Indian_Climate_Dataset_2024_2025.csv')

# Display data types and non-null values
print("DataFrame Info:")
display(df.info())

# Check for missing values
print("\nMissing Values:")
display(df.isnull().sum())

# Check for duplicate rows
print(f"\nNumber of duplicate rows: {df.duplicated().sum()}")
```

```python
# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' as the index for time-series analysis
df.set_index('Date', inplace=True)

# Display the DataFrame information again to confirm changes
print("DataFrame Info after Date conversion and index setting:")
display(df.info())
display(df.head())
```

## 2. Trend Analysis

```python
# Calculate daily averages for numerical columns across all cities
daily_avg_df = df.resample('D').mean(numeric_only=True)

# Plotting trends for key climate parameters
plt.figure(figsize=(18, 12))

# Trend for Average Temperature
plt.subplot(3, 1, 1)
sns.lineplot(data=daily_avg_df, x=daily_avg_df.index, y='Temperature_Avg (°C)')
plt.title('Daily Average Temperature Trend Across All Cities')
plt.ylabel('Average Temperature (°C)')
plt.xlabel('')
plt.grid(True)

# Trend for Rainfall
plt.subplot(3, 1, 2)
sns.lineplot(data=daily_avg_df, x=daily_avg_df.index, y='Rainfall (mm)')
plt.title('Daily Average Rainfall Trend Across All Cities')
plt.ylabel('Average Rainfall (mm)')
plt.xlabel('')
plt.grid(True)

# Trend for AQI
plt.subplot(3, 1, 3)
sns.lineplot(data=daily_avg_df, x=daily_avg_df.index, y='AQI')
plt.title('Daily Average AQI Trend Across All Cities')
plt.ylabel('Average AQI')
plt.xlabel('Date')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## 3. City Comparison

```python
# Group by City and calculate the mean for relevant climate parameters
city_comparison_df = df.groupby('City')[['Temperature_Avg (°C)', 'Rainfall (mm)', 'AQI', 'Humidity (%)', 'Wind_Speed (km/h)']].mean().sort_values(by='Temperature_Avg (°C)', ascending=False)

display(city_comparison_df)

# Plotting city-wise comparison for key parameters
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 12))
fig.suptitle('City-wise Comparison of Key Climate Parameters', fontsize=16)

# Average Temperature
sns.barplot(x=city_comparison_df.index, y='Temperature_Avg (°C)', data=city_comparison_df, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Average Temperature by City')
axes[0, 0].set_ylabel('Avg Temperature (°C)')
axes[0, 0].set_xlabel('')
axes[0, 0].tick_params(axis='x', rotation=45)

# Average Rainfall
sns.barplot(x=city_comparison_df.index, y='Rainfall (mm)', data=city_comparison_df, ax=axes[0, 1], palette='Blues')
axes[0, 1].set_title('Average Rainfall by City')
axes[0, 1].set_ylabel('Avg Rainfall (mm)')
axes[0, 1].set_xlabel('')
axes[0, 1].tick_params(axis='x', rotation=45)

# Average AQI
sns.barplot(x=city_comparison_df.index, y='AQI', data=city_comparison_df, ax=axes[1, 0], palette='Reds')
axes[1, 0].set_title('Average AQI by City')
axes[1, 0].set_ylabel('Avg AQI')
axes[1, 0].set_xlabel('')
axes[1, 0].tick_params(axis='x', rotation=45)

# Average Humidity
sns.barplot(x=city_comparison_df.index, y='Humidity (%)', data=city_comparison_df, ax=axes[1, 1], palette='Greens')
axes[1, 1].set_title('Average Humidity by City')
axes[1, 1].set_ylabel('Avg Humidity (%)')
axes[1, 1].set_xlabel('')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## 4. Correlation Analysis

```python
# Select numerical columns for correlation analysis
correlation_cols = [
    'Temperature_Max (°C)', 'Temperature_Min (°C)', 'Temperature_Avg (°C)',
    'Humidity (%)', 'Rainfall (mm)', 'Wind_Speed (km/h)', 'Pressure (hPa)',
    'Cloud_Cover (%)', 'AQI'
]

# Calculate the correlation matrix
correlation_matrix = df[correlation_cols].corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Climate Parameters')
plt.show()

display(correlation_matrix)
```

## 5. Actionable Insights and Conclusion

### Actionable Insights:

*   **Temperature Trends:** The trend analysis shows fluctuations in average temperature. Cities and regions experiencing higher average temperatures could benefit from early warning systems for heatwaves and public health advisories.
*   **Rainfall Patterns:** Monitoring rainfall trends is crucial for water resource management and agriculture. Areas with unpredictable or heavy rainfall patterns might require improved drainage systems and flood preparedness.
*   **Air Quality Index (AQI):** Cities with consistently high AQI values should prioritize implementing stricter pollution control measures, promoting public transportation, and encouraging green initiatives to improve air quality and public health.
*   **City-Specific Interventions:** The city comparison highlights variations in climate parameters. Tailored strategies based on specific city profiles can be developed, e.g., for cities with high average temperatures, focus on urban greening and cool roof initiatives.
*   **Humidity Management:** High humidity levels can impact comfort and lead to health issues. Strategies to manage indoor and outdoor humidity, especially in high-humidity cities, could be beneficial.
*   **Wind Speed and Cloud Cover:** These factors influence local climate conditions. Understanding their patterns can help in urban planning, such as optimizing building ventilation and utilizing renewable energy sources like wind power.
*   **Correlation between Parameters:** The correlation matrix provides insights into how different climate factors relate. For instance, understanding the correlation between temperature, humidity, and AQI can help predict extreme weather events and air pollution episodes more effectively.
